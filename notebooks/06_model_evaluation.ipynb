```python
import pandas as pd
import joblib

from sklearn.metrics import (
    classification_report,
    confusion_matrix
)
```

```python
df = pd.read_csv(
    "../outputs/cleaned_tickets.csv"
)

df.head()
```

```python
model = joblib.load(
    "../models/category_model.pkl"
)

vectorizer = joblib.load(
    "../models/vectorizer.pkl"
)
```

```python
X = vectorizer.transform(df["text"])

y_true = df["Category"]
```

```python
y_pred = model.predict(X)
```

```python
print(
    classification_report(
        y_true,
        y_pred
    )
)
```
