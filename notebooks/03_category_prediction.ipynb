```python
import joblib

# Load saved model
model = joblib.load(
    "../models/category_model.pkl"
)

# Load vectorizer
vectorizer = joblib.load(
    "../models/vectorizer.pkl"
)

print("Model loaded successfully")
```

```python
import re

def clean_text(text):

    # Convert to string
    text = str(text)

    # Replace line breaks, tabs
    text = text.replace("\n", " ")
    text = text.replace("\r", " ")
    text = text.replace("\t", " ")

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()


def predict_category(subject, description, account):

    # Clean inputs
    subject = clean_text(subject)
    description = clean_text(description)
    account = clean_text(account)

    # Combine text
    text = f"{subject} {description} {account}"

    # Convert text into vector
    vector = vectorizer.transform([text])

    # Predict category
    prediction = model.predict(vector)

    return prediction[0]
```
