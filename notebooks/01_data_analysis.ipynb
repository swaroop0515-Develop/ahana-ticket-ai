```python
import pandas as pd


import re

def clean_ticket_text(text):

    text = str(text)

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+', ' ', text)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', ' ', text)

    # Remove IP addresses
    text = re.sub(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', ' ', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    return text.strip()

    
# Load Excel file
df = pd.read_excel(
    "../data/cleaned_tickets.xlsx",
    skiprows=8
)

# Show first rows
df.head()



required_columns = [
    "Subject",
    "Description",
    "Account",
    "Impact",
    "Urgency",
    "Priority",
    "Category",
    "Subcategory",
    "Group",
    "Request Type",
    "Environment",
    "Environment Type"
]
category_mapping = {

    "Application": "Applications",
    "Software": "Applications",
    "Google": "Applications",
    "Microsoft": "Applications",

    "Server": "Servers and Systems",

    "CrowdStrike": "Security",

    "Scanner/Printer": "Hardware",

    "Unknown": "Not Applicable"
}

df["Category"] = df["Category"].replace(category_mapping)
print(df["Category"].value_counts())

df = df[required_columns]

df.head()

df["text"] = (
    df["Subject"].apply(clean_ticket_text)
    + " "
    + df["Description"].apply(clean_ticket_text)
    + " "
    + df["Account"].apply(clean_ticket_text)
)

df[["text", "Category"]].head()

df.to_csv(
    "../outputs/cleaned_tickets.csv",
    index=False
)

print("Cleaned data saved successfully")
```
